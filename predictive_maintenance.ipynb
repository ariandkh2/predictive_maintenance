<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc" size= 10> ساختار داده‌ها
</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=4>
    داده‌ها در این پروژه در قالب یک فایل CSV ذخیره شده‌اند که شامل اطلاعات مختلفی در مورد دستگاه کمپرسور و ویژگی‌های عملکردی آن است. این فایل به صورت جدول داده‌ها (DataFrame) بارگذاری می‌شود که هر ستون نشان‌دهنده یک ویژگی خاص است. به طور کلی، ساختار داده‌ها شامل ستون‌های زیر است:
    <ul style="list-style-position: inside; text-align:right;">
        <code>کد بررسی</code>: شناسه یا کد منحصر به فرد برای هر بررسی انجام‌شده روی دستگاه.
        <li><code>دمای هوای اطراف ماشین (کلوین)</code>: دمای محیط یا هوای اطراف دستگاه کمپرسور که می‌تواند تأثیر زیادی بر روی عملکرد آن داشته باشد.
        <li><code>دمای فرآیند (کلوین)</code>: دمای فرآیند داخل دستگاه که نشان‌دهنده دمای عملیاتی دستگاه است.
        <li><code>سرعت چرخش اجزای ماشین (دور در دقیقه)</code>: سرعت چرخش قطعات مختلف دستگاه که می‌تواند به عنوان معیاری برای شناسایی بار اضافی یا خرابی در قطعات باشد.
        <li><code>گشتاور ماشین (نیوتون متر)</code>: میزان گشتاوری که دستگاه تولید می‌کند که اطلاعاتی در مورد عملکرد دستگاه و بار آن فراهم می‌آورد.
        <li><code>مدت زمان سایش ابزار (دقیقه)</code>: مدت زمانی که ابزارهای دستگاه در حال استفاده بوده‌اند و می‌تواند ارتباطی با خرابی‌ها یا نیاز به تعمیرات داشته باشد.
        <li><code>وضعیت خرابی</code>: وضعیت کلی دستگاه (0 برای بدون خرابی و 1 برای خرابی ماشین).
        <li><code>نوع خرابی</code>: نوع خرابی دستگاه، مانند خرابی در اتلاف گرما، خرابی نیرو، خرابی ناشی از فشار بیش‌ازحد، یا خرابی ناشی از ساییدگی ابزار.
    </ul>
    <br>
    با استفاده از این داده‌ها، می‌توانیم تحلیل‌هایی مانند شناسایی الگوهای خرابی در زمان‌های مختلف، بررسی تاثیر شرایط محیطی و عملیاتی دستگاه بر خرابی‌ها، و پیش‌بینی خرابی‌های احتمالی آینده انجام دهیم. به علاوه، تحلیل این داده‌ها می‌تواند به بهینه‌سازی فرآیندهای نگهداری و تعمیرات کمک کند.
</font>
</p>


```python
import pandas as pd
import requests
from io import StringIO

# لینک به فایل Google Drive
file_url = "https://drive.google.com/file/d/1oPWP0P52vH3cUG_922hvLElpHy3lIqvO/view?usp=drive_link"

# استخراج ID فایل از لینک
file_id = file_url.split('/d/')[1].split('/')[0]

# ساخت لینک مستقیم دانلود
direct_url = f"https://drive.google.com/uc?id={file_id}&export=download"

# دانلود فایل و بارگذاری به pandas
response = requests.get(direct_url)

# خواندن محتوا به عنوان CSV
data = pd.read_csv(StringIO(response.text))

# نمایش داده‌ها به‌صورت جدول
data.head()
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
    <font face="vazir" color="#0099cc" size=10>پاکسازی داده‌ها</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
    <font face="vazir" size=4>
        پاکسازی داده‌ها یکی از مراحل کلیدی برای اطمینان از کیفیت داده‌ها و دقت تحلیل‌ها است. در این پروژه، مراحل زیر برای پاکسازی داده‌ها انجام شده است:
        <ul style="list-style-position: inside; text-align:right;">
            <li><code>استانداردسازی نام ستون‌ها:</code> حذف فاصله‌های اضافی از نام ستون‌ها و اطمینان از یکنواختی نام‌گذاری برای جلوگیری از خطاهای احتمالی.</li>
            <li><code>جایگزینی مقادیر عددی با توضیحات متنی:</code> مقادیر عددی ستون "وضعیت خرابی" به عبارات متنی "بدون خرابی" و "خرابی ماشین" تبدیل شدند.</li>
            <li><code>ترجمه مقادیر نوع خرابی:</code> مقادیر ستون "نوع خرابی" از انگلیسی به فارسی ترجمه شدند تا خوانایی بیشتری داشته باشند. به عنوان مثال:
                <ul style="list-style-position: inside; text-align:right;">
                    <li>"No Failure" به "بدون خرابی" تبدیل شد.</li>
                    <li>"Heat Dissipation Failure" به "خرابی دفع گرما" تبدیل شد.</li>
                    <li>"Power Failure" به "خرابی توان" تبدیل شد.</li>
                </ul>
            </li>
            <li><code>تبدیل داده‌ها به نوع عددی:</code> ستون‌هایی که مقادیر عددی داشتند، بررسی و به نوع عددی تبدیل شدند تا برای محاسبات آماری آماده باشند.</li>
        </ul>
        <br>
        با انجام این فرآیندها، داده‌ها برای تحلیل‌های دقیق‌تر آماده شدند و امکان تمرکز بر روی نتایج اصلی فراهم شد.
    </font>
</p>


```python
import numpy as np
import pandas as pd

# استانداردسازی نام ستون‌ها (حذف فاصله‌های اضافی از نام ستون‌ها)
data.columns = data.columns.str.strip()

# تغییر مقادیر ستون "وضعیت خرابی" به عبارات توصیفی
data["وضعیت خرابی"] = np.where(data["وضعیت خرابی"] == 0, "بدون خرابی", "خرابی ماشین")

# ترجمه مقادیر ستون "نوع خرابی" به فارسی
data["نوع خرابی"] = data["نوع خرابی"].replace({
    "No Failure": "بدون خرابی",
    "Heat Dissipation Failure": "خرابی دفع گرما",
    "Power Failure": "خرابی توان",
    "Overstrain Failure": "خرابی ناشی از فشار بیش‌ازحد",
    "Tool Wear Failure": "خرابی ابزار",
    "Random Failures": "خرابی‌های تصادفی" 
})

# تبدیل ستون‌های عددی به نوع عددی (در صورت نیاز به اجتناب از خطاهای تبدیل، استفاده از pd.to_numeric)
numeric_columns = ["دمای هوای اطراف ماشین (کلوین)", "دمای فرآیند (کلوین)", 
                   "سرعت چرخش اجزای ماشین (دور در دقیقه)", "گشتاور ماشین (نیوتون متر)", 
                   "مدت زمان سایش ابزار (دقیقه)"]

data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')

# نمایش پیش‌نمایش از داده‌های تمیز شده
data.head()
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
    <font face="vazir" color="#0099cc" size=10>تعداد تکرار خرابی ها</font>
</h2>

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
    <font face="vazir" size=4>
        در این بخش، تحلیل داده‌ها را با بررسی تعداد تکرار هر نوع خرابی آغاز می‌کنیم. این تحلیل به ما کمک می‌کند تا متوجه شویم کدام نوع خرابی بیشتر رخ داده و اولویت‌بندی در بررسی‌ها و تحلیل‌های عمیق‌تر را مشخص کنیم.  
        <br><br>
        برای این منظور، تعداد تکرار هر خرابی به صورت نمودار نمایش داده می‌شود تا بتوانیم مقایسه‌ای بصری میان انواع خرابی‌ها انجام دهیم.
    </font>
</p>


```python
import warnings
import arabic_reshaper
from bidi.algorithm import get_display
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager

# غیرفعال کردن نمایش پیغام‌های هشدار
warnings.filterwarnings('ignore', category=FutureWarning)

# تنظیم فونت
font_path = 'C:/Windows/Fonts/Tahoma.ttf'  # مسیر فونت فارسی
prop = font_manager.FontProperties(fname=font_path)

# اصلاح متن فارسی برای نمایش صحیح
def reshaped_text(text):
    reshaped = arabic_reshaper.reshape(text)
    return get_display(reshaped)

# فیلتر کردن داده‌ها تا فقط خرابی‌های ماشین نمایش داده شوند
filtered_data = data[data['نوع خرابی'] != 'بدون خرابی']

# گروه‌بندی داده‌ها بر اساس نوع خرابی و شمارش تعداد هر نوع خرابی
failure_counts = filtered_data['نوع خرابی'].value_counts()

# مرتب‌سازی داده‌ها طبق تعداد خرابی‌ها به ترتیب نزولی
failure_counts = failure_counts.sort_values(ascending=False)

# رسم نمودار میله‌ای
plt.figure(figsize=(10,6))
ax = sns.barplot(x=failure_counts.index, y=failure_counts.values, palette='viridis')

# اعمال فونت فارسی و اصلاح متن برای عنوان و برچسب‌ها
plt.title(reshaped_text('تعداد هر نوع خرابی'), fontsize=16, fontproperties=prop)
plt.xlabel(reshaped_text('نوع خرابی'), fontsize=12, fontproperties=prop)
plt.ylabel(reshaped_text('تعداد خرابی‌ها'), fontsize=12, fontproperties=prop)

# اصلاح ترتیب برچسب‌های محور X و نمایش حروف فارسی
ticks = range(len(failure_counts))
plt.xticks(ticks, [reshaped_text(label) for label in failure_counts.index], rotation=45, ha='right', fontproperties=prop)

# اضافه کردن دیتالیبل به هر میله
for p in ax.patches:
    ax.annotate(f'{p.get_height():,.0f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                fontsize=12, color='black', 
                xytext=(0, 8), textcoords='offset points')

# اصلاح چیدمان و نمایش نمودار
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bidi.algorithm import get_display
import arabic_reshaper
from matplotlib import font_manager

# تنظیم فونت فارسی
font_path = 'C:/Windows/Fonts/Tahoma.ttf'  # مسیر فونت فارسی
prop = font_manager.FontProperties(fname=font_path)

# اصلاح متن فارسی
def reshaped_text(text):
    reshaped = arabic_reshaper.reshape(text)
    return get_display(reshaped)

# فیلتر کردن سطرهای مربوط به خرابی دفع گرما
heat_failure_data = data[data["نوع خرابی"] == "خرابی دفع گرما"]

# انتخاب ستون‌های خاص برای محاسبه میانگین (مثلاً: دمای هوای اطراف ماشین، دمای فرآیند و ...)
selected_columns = ["دمای هوای اطراف ماشین (کلوین)", "دمای فرآیند (کلوین)", 
                    "سرعت چرخش اجزای ماشین (دور در دقیقه)", "گشتاور ماشین (نیوتون متر)", 
                    "مدت زمان سایش ابزار (دقیقه)"]

# فیلتر کردن داده‌ها برای ستون‌های خاص
selected_data = heat_failure_data[selected_columns]

# محاسبه میانگین برای ستون‌های انتخاب شده در وضعیت خرابی دفع گرما
heat_failure_mean = selected_data.mean()

# محاسبه میانگین برای ستون‌های انتخاب شده در تمام داده‌ها
overall_mean = data[selected_columns].mean()

# محاسبه تفاوت میانگین خرابی دفع گرما نسبت به میانگین کل
difference = heat_failure_mean - overall_mean

# تبدیل به DataFrame برای رسم نمودار
df_diff = pd.DataFrame({
    'میانگین خرابی دفع گرما': heat_failure_mean,
    'میانگین کل': overall_mean,
    'تفاوت': difference
})

# تنظیمات نمودار
plt.figure(figsize=(12, 6))  # افزایش اندازه طولی نمودار

# رسم نمودار میله‌ای برای مقایسه
df_diff[['میانگین خرابی دفع گرما', 'میانگین کل']].plot(kind='bar', ax=plt.gca(), color=['skyblue', 'lightgreen'], edgecolor='black')

# اضافه کردن دیتالیبل
for i in range(len(df_diff)):
    # فاصله بیشتری برای دیتالیبل‌ها در نظر می‌گیریم
    plt.text(i - 0.1, df_diff['میانگین خرابی دفع گرما'][i] + 0.1, f"{df_diff['میانگین خرابی دفع گرما'][i]:.2f}", ha='center', va='bottom', fontproperties=prop)
    plt.text(i + 0.1, df_diff['میانگین کل'][i] + 0.1, f"{df_diff['میانگین کل'][i]:.2f}", ha='center', va='bottom', fontproperties=prop)

# اصلاح ترتیب برچسب‌های محور افقی
ticks = range(len(df_diff))
plt.xticks(ticks, [reshaped_text(label) for label in df_diff.index], rotation=45, ha='right', fontproperties=prop)

# عنوان و برچسب‌ها
plt.title(reshaped_text("مقایسه میانگین مقادیر خرابی دفع گرما نسبت به میانگین کل"), fontsize=16, fontproperties=prop)
plt.xlabel(reshaped_text("ویژگی‌ها"), fontsize=12, fontproperties=prop)
plt.ylabel(reshaped_text("مقدار میانگین"), fontsize=12, fontproperties=prop)

# اصلاح ترتیب حروف در لجند
plt.legend([reshaped_text('میانگین خرابی دفع گرما'), reshaped_text('میانگین کل')], prop=prop)

# چیدمان و نمایش نمودار
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bidi.algorithm import get_display
import arabic_reshaper

# ایجاد بازه‌ها و برچسب‌ها
speed_bins = np.arange(1160, 2890, 100)  # بازه‌های 100 واحدی
speed_labels = [f"{b}-{b + 100}" for b in speed_bins[:-1]]

# ایجاد ستون بازه‌های سرعت چرخش
data["سرعت بازه"] = pd.cut(data["سرعت چرخش اجزای ماشین (دور در دقیقه)"], bins=speed_bins, labels=speed_labels, right=False)

# شمارش تعداد کل بررسی‌ها در هر بازه
speed_counts = data["سرعت بازه"].value_counts(sort=False)

# شمارش تعداد خرابی‌های دفع گرما در هر بازه
heat_failure_speed_counts = data[data["نوع خرابی"] == "خرابی دفع گرما"]["سرعت بازه"].value_counts(sort=False)

# اعمال تنظیمات برای پشتیبانی از متن فارسی
plt.rcParams['font.family'] = 'Tahoma'

# رسم نمودار
fig, ax1 = plt.subplots(figsize=(16, 6))  # افزایش سایز نمودار

# نمودار میله‌ای برای تعداد کل بررسی‌ها
bars = ax1.bar(speed_labels, speed_counts, color="lightgreen", edgecolor="black", label=get_display(arabic_reshaper.reshape("تعداد کل بررسی‌ها")))
ax1.set_ylabel(get_display(arabic_reshaper.reshape("تعداد کل بررسی‌ها")), fontsize=12, color="green")
ax1.set_xlabel(get_display(arabic_reshaper.reshape("بازه‌های سرعت چرخش اجزای ماشین (دور در دقیقه)")), fontsize=12)
ax1.set_title(get_display(arabic_reshaper.reshape("تعداد بررسی‌ها و خرابی‌های دفع گرما در بازه‌های سرعت چرخش")), fontsize=14)
ax1.tick_params(axis="y", labelcolor="green")

# تنظیم چرخش نوشته‌های محور افقی
plt.xticks(rotation=45, ha='right', fontsize=10)

# ایجاد محور دوم برای نمایش تعداد خرابی دفع گرما
ax2 = ax1.twinx()
heat_failure_speed_counts_reindexed = heat_failure_speed_counts.reindex(speed_labels, fill_value=0)
line = ax2.plot(speed_labels, heat_failure_speed_counts_reindexed, color="orange", marker="o", label=get_display(arabic_reshaper.reshape("تعداد خرابی دفع گرما")), lw=2)
ax2.set_ylabel(get_display(arabic_reshaper.reshape("تعداد خرابی دفع گرما")), fontsize=12, color="orange")
ax2.tick_params(axis="y", labelcolor="orange")
ax2.set_ylim(0, max(heat_failure_speed_counts_reindexed) * 1.2)  # تنظیم بازه مناسب محور دوم

# افزودن دیتالجند
lines_labels = [bars[0], line[0]]
labels = [get_display(arabic_reshaper.reshape("تعداد کل بررسی‌ها")), get_display(arabic_reshaper.reshape("تعداد خرابی دفع گرما"))]
ax1.legend(lines_labels, labels, loc="upper left", fontsize=10)

# افزودن دیتالیبل‌ها
for bar in bars:
    ax1.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height()}', ha='center', va='bottom', fontsize=10)

for x, y in zip(speed_labels, heat_failure_speed_counts_reindexed):
    ax2.text(x, y, f'{int(y)}', color="orange", fontsize=10, ha='center', va='bottom')

# نمایش نمودار
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# محاسبه بازه‌ها برای هر پارامتر
def calculate_range(data, column_name):
    return f"{data[column_name].min()} - {data[column_name].max()}"

# گروه‌بندی داده‌ها بر اساس نوع خرابی
grouped_data = data.groupby('نوع خرابی')

# ایجاد یک لیست برای ذخیره بازه‌ها
rows = []

# محاسبه بازه‌ها برای هر خرابی
for name, group in grouped_data:
    row = {
        "نوع خرابی": name,
        "مدت زمان سایش ابزار": calculate_range(group, "مدت زمان سایش ابزار (دقیقه)"),
        "گشتاور ماشین": calculate_range(group, "گشتاور ماشین (نیوتون متر)"),
        "سرعت چرخش اجزای ماشین": calculate_range(group, "سرعت چرخش اجزای ماشین (دور در دقیقه)"),
        "دمای هوای اطراف ماشین": calculate_range(group, "دمای هوای اطراف ماشین (کلوین)"),
        "دمای فرآیند": calculate_range(group, "دمای فرآیند (کلوین)")
    }
    rows.append(row)

# تبدیل لیست به DataFrame
range_table = pd.DataFrame(rows)

# نمایش جدول بازه‌ها
range_table
```

    برای نمایش بازه‌های عددی هر پارامتر برای هر نوع خرابی به گونه‌ای که مقایسه آن‌ها ساده و کارآمد باشد، بهترین گزینه نمودار Boxplot (نمودار جعبه‌ای) است. این نمودار به‌طور خلاصه اطلاعات آماری مانند کمینه، بیشینه، میانه، و چارک‌ها را برای هر دسته داده (در اینجا انواع خرابی) نمایش می‌دهد و بسیار مناسب برای مقایسه بازه‌های داده‌ای بین گروه‌ها است.
    <br><br>
    دلیل انتخاب Boxplot:
    <br>
    مقایسه‌ی بصری ساده:
    <br>
    می‌توانید تمام انواع خرابی را در کنار هم نمایش دهید و بازه‌های عددی هر پارامتر را مقایسه کنید.
    <br><br>
    اطلاعات بیشتر در مقایسه با جدول:
    <br>
    علاوه بر بازه (کمینه و بیشینه)، این نمودار میانه و چارک‌ها را نیز نشان می‌دهد.
    <br><br>
    نمایش داده‌های پرت:
    <br>
    نقاط پرت که ممکن است روی تحلیل تأثیر بگذارند به‌وضوح در این نمودار مشخص می‌شوند. این نقاط به‌عنوان "outliers" در نمودار جعبه‌ای به‌صورت نقاط جداگانه (به‌طور معمول با علامت ستاره یا دایره) نمایش داده می‌شوند.
    <br><br>
    در نتیجه، نمودار جعبه‌ای یک ابزار بصری مفید برای مقایسه و تحلیل داده‌ها است که علاوه بر نمایش بازه‌ها، میانه‌ها، و چارک‌ها، امکان شناسایی داده‌های پرت را نیز فراهم می‌کند.
    </font>

</h2>

<p dir="rtl" style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
    <font face="vazir" size=4>
    نمودار جعبه‌ای (Boxplot) یک ابزار گرافیکی است که برای نمایش توزیع داده‌ها و ویژگی‌های آماری آن‌ها استفاده می‌شود. در این نمودار، اطلاعات مهمی مانند میانه، چارک‌ها، و مقادیر پرت برای هر دسته داده به نمایش درمی‌آید.
    <br><br>
    اجزای اصلی نمودار جعبه‌ای عبارتند از:
    <br><br>
    1. <b>جعبه (Box):</b>  
    <br>
    جعبه اصلی نمودار که از دو خط افقی به نام "چارک اول" (Q1) و "چارک سوم" (Q3) تشکیل شده است. جعبه نمایانگر 50% مرکزی داده‌ها است که بین این دو چارک قرار دارند.
    <br><br>
    2. <b>خط میانه (Median):</b>  
    <br>
    خطی که به داخل جعبه رسم می‌شود و میانه داده‌ها (Q2) را نشان می‌دهد. این خط نمایانگر نقطه‌ای است که داده‌ها را به دو بخش برابر تقسیم می‌کند.
    <br><br>
    3. <b>بازه بین چارک‌ها (Interquartile Range - IQR):</b>  
    <br>
    فاصله بین چارک اول (Q1) و چارک سوم (Q3) که نشان‌دهنده پخش شدن داده‌ها است. این بازه 50 درصد از داده‌ها را در بر می‌گیرد.
    <br><br>
    4. <b>شاخ‌ها (Whiskers):</b>  
    <br>
    خطوط افقی که از دو طرف جعبه خارج می‌شوند و نشان‌دهنده پراکندگی داده‌ها خارج از چارک‌ها هستند. طول این شاخ‌ها معمولاً تا 1.5 برابر فاصله IQR از Q1 و Q3 است.
    <br><br>
    5. <b>داده‌های پرت (Outliers):</b>  
    <br>
    نقاطی که به‌طور واضح از سایر داده‌ها فاصله دارند و خارج از محدوده‌ی شاخ‌ها قرار می‌گیرند. این نقاط معمولاً به‌صورت دایره یا ستاره در نمودار نشان داده می‌شوند.
    <br><br>
    <b>چرا استفاده از نمودار جعبه‌ای مفید است؟</b>
    <br><br>
    - <b>تحلیل توزیع داده‌ها:</b> این نمودار به‌طور سریع و واضح نشان می‌دهد که داده‌ها چگونه توزیع شده‌اند و آیا تقارن دارند یا خیر.
    <br><br>
    - <b>تشخیص داده‌های پرت:</b> می‌توانید نقاط غیرعادی یا پرت را شناسایی کنید که ممکن است نیاز به بررسی بیشتر داشته باشند.
    <br><br>
    - <b>مقایسه گروه‌ها:</b> اگر چندین مجموعه داده داشته باشید، نمودار جعبه‌ای به شما این امکان را می‌دهد که توزیع و ویژگی‌های مختلف هر گروه را با هم مقایسه کنید.
    </font>
</p>
